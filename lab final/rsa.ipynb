### Choose prime number

```python
p=11
q=13
```

### Calculate modulas n

```python
n=p*q
n
```

### Calculate euler totient function

```python
phi=(p-1)*(q-1)
phi
```

### choose a value which is phi's common factor

```python
e=7
```

### Calculate Private Key

```python
d=pow(e,-1,phi)
d
```

```python
print(f"Public Key: (e={e}, n={n})")
print(f"Private Key: (d={d}, n={n})")
```
