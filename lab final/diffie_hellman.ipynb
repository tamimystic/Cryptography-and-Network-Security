## Primitive kew Generator

```python
p=int(input("Enter a Prime number (public key) : "))
```

```python
list_primitive=[]

for i in range(1,p):
    is_primitive=True

    for j in range(1,p-1):
        if pow(i,j,p)==1:
            is_primitive=False
            break
    if is_primitive:
        list_primitive.append(i)

list_primitive
```
