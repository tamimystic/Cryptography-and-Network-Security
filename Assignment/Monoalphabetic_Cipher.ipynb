Write a program to implement a Monoalphabetic Substitution Cipher. <br>

Your program should: <br>

Use a fixed substitution key for A–Z. <br>
Encrypt a plaintext message. <br>
Decrypt the ciphertext using the inverse substitution table. <br>
Show both the substitution table and the result.<br>

# plaintext input

```python
plaintext=input("Enter your message: ").upper()
```

```python
print(plaintext)
```

# fixed substitution key for A–Z

```python
alphabet="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
substitution="QWERTYUIOPASDFGHJKLZXCVBNM"
```

# show

```python
print("Substitution Table")
print("Original   :",alphabet)
print("Substitute :",substitution)
```

# encryption

```python
ciphertext=""

for char in plaintext:
    if char.isalpha():
        index=alphabet.find(char)
        ciphertext=ciphertext+substitution[index]
    else:
        ciphertext=ciphertext+char

print("Encrypted Message:",ciphertext)
```

# decryption

```python
decrypted_text=""

for char in ciphertext:
    if char.isalpha():
        index=substitution.find(char)
        decrypted_text=decrypted_text+alphabet[index]
    else:
        decrypted_text=decrypted_text+char

print("Decrypted Message:",decrypted_text)
```
