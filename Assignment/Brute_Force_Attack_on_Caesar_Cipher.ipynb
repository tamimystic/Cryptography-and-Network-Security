# brute force function

```python
def brute_force_caesar(text):
    for key in range(1,26):
        decrypted_text=""

        for char in text:
            if char.isalpha():
                if char.islower():
                    shifted=((ord(char)-ord('a')-key)%26)+ord('a')
                else:
                    shifted=((ord(char)-ord('A')-key)%26)+ord('A')

                decrypted_text+=chr(shifted)
            else:
                decrypted_text+=char

        print("Key",key,":",decrypted_text)
```
