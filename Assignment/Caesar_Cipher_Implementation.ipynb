Write a program to implement the Caesar Cipher. <br>

Your program should: <br>

Take a plaintext message as input. <br>
Take a key value between 1 and 25.<br>
Encrypt the plaintext using Caesar Cipher. <br>
Decrypt the ciphertext back to plaintext. <br>

# user input

```python
plaintext=input("Enter your message: ")
key=int(input("Enter key (1 to 25): "))
```

```python
print(plaintext)
print(key)
```

# encryption function

```python
def encrypt_caesar(text,key):
    encrypted_text=""

    for char in text:
        if char.isalpha():
            if char.islower():
                shifted=((ord(char)-ord('a')+key)%26)+ord('a')
            else:
                shifted=((ord(char)-ord('A')+key)%26)+ord('A')

            encrypted_text+=chr(shifted)
        else:
            encrypted_text+=char

    return encrypted_text
```

# decryption function

```python
def decrypt_caesar(text,key):
    decrypted_text=""

    for char in text:
        if char.isalpha():
            if char.islower():
                shifted=((ord(char)-ord('a')-key)%26)+ord('a')
            else:
                shifted=((ord(char)-ord('A')-key)%26)+ord('A')

            decrypted_text+=chr(shifted)
        else:
            decrypted_text+=char

    return decrypted_text
```

# encrypt the plaintext

```python
ciphertext=encrypt_caesar(plaintext,key)
print("Encrypted Message:",ciphertext)
```

# decrypt the ciphertext

```python
decrypted_text=decrypt_caesar(ciphertext,key)
print("Decrypted Message:",decrypted_text)
```
