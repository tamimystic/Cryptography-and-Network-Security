# Diffie-Hellman Key Exchange Implementation

## Tasks

1. Take a prime number as input
2. Take a primitive root or generator as input
3. Take Alice’s private key as input
4. Take Bob’s private key as input
5. Generate Alice’s public key
6. Generate Bob’s public key
7. Generate the shared secret key for Alice
8. Generate the shared secret key for Bob
9. Verify whether both shared secret keys are equal
10. Display the public keys and the shared secret keys

Input Values

```python
p=int(input("Enter prime number: "))
g=int(input("Enter primitive root: "))

alice_private=int(input("Enter Alice's private key: "))
bob_private=int(input("Enter Bob's private key: "))
```

```python
print(alice_private)
print(bob_private)
```

Generate Alice's Public Key

```python
alice_public=1

for i in range(alice_private):
    alice_public=(alice_public*g)%p

print("Alice Public Key =",alice_public)
```

Generate Bob's Public Key

```python
bob_public=1

for i in range(bob_private):
    bob_public=(bob_public*g)%p

print("Bob Public Key =",bob_public)
```

Generate Shared Secret Key for Alice

```python
alice_secret=1

for i in range(alice_private):
    alice_secret=(alice_secret*bob_public)%p

print("Alice Shared Secret Key =",alice_secret)
```

Generate Shared Secret Key for Bob

```python
bob_secret=1

for i in range(bob_private):
    bob_secret=(bob_secret*alice_public)%p

print("Bob Shared Secret Key =",bob_secret)
```

Verify Shared Secret Keys

```python
if alice_secret==bob_secret:
    print("Both shared secret keys are equal")
else:
    print("Shared secret keys are not equal")
```
