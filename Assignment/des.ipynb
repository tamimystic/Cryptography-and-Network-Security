## Tasks

Input Plaintext Block

```python
plaintext=input("Enter 64-bit binary plaintext: ")
plaintext
```

Initial Permutation Table

```python
ip_table=[
58,50,42,34,26,18,10,2,
60,52,44,36,28,20,12,4,
62,54,46,38,30,22,14,6,
64,56,48,40,32,24,16,8,
57,49,41,33,25,17,9,1,
59,51,43,35,27,19,11,3,
61,53,45,37,29,21,13,5,
63,55,47,39,31,23,15,7
]
```

Apply Initial Permutation

```python
ip_output=""

for position in ip_table:
    ip_output=ip_output+plaintext[position-1]

print("After Initial Permutation:")
print(ip_output)
```

Split into Left and Right Halves

```python
left=ip_output[:32]
right=ip_output[32:]

print("Left Half:")
print(left)

print("Right Half:")
print(right)
```

Expansion Permutation Table

```python
expansion_table=[
32,1,2,3,4,5,
4,5,6,7,8,9,
8,9,10,11,12,13,
12,13,14,15,16,17,
16,17,18,19,20,21,
20,21,22,23,24,25,
24,25,26,27,28,29,
28,29,30,31,32,1
]
```

Expansion Permutation

```python
expanded=""

for position in expansion_table:
    expanded=expanded+right[position-1]

print("Expanded Right Half:")
print(expanded)
```

Input Round Key

```python
round_key=input("Enter 48-bit round key: ")
round_key
```

XOR Operation

```python
xor_result=""

for i in range(48):

    if expanded[i]==round_key[i]:
        xor_result=xor_result+"0"
    else:
        xor_result=xor_result+"1"

print("After XOR:")
print(xor_result)
```

S-Box 1

```python
sbox1=[
[14,4,13,1,2,15,11,8,3,10,6,12,5,9,0,7],
[0,15,7,4,14,2,13,1,10,6,12,11,9,5,3,8],
[4,1,14,8,13,6,2,11,15,12,9,7,3,10,5,0],
[15,12,8,2,4,9,1,7,5,11,3,14,10,0,6,13]
]
```

S-Box Substitution

```python
block=xor_result[:6]

row_binary=block[0]+block[5]
col_binary=block[1]+block[2]+block[3]+block[4]

row=int(row_binary,2)
col=int(col_binary,2)

value=sbox1[row][col]

binary_output=bin(value)[2:]

while len(binary_output)<4:
    binary_output="0"+binary_output

print("S-Box Input :",block)
print("S-Box Output:",binary_output)
```
