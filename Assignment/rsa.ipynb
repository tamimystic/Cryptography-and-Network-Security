# RSA Key Generation, Encryption, and Decryption

## Tasks

1. Take two prime numbers p and q as input
2. Calculate n
3. Calculate Euler’s totient function φ(n)
4. Take or choose a public key e such that gcd(e, φ(n)) = 1
5. Calculate the private key d
6. Take a numeric plaintext message M as input, where M < n
7. Encrypt the plaintext message
8. Decrypt the ciphertext message
9. Display the public key, private key, ciphertext, and decrypted plaintext

Input Prime Numbers

```python
p=int(input("Enter prime number p: "))
q=int(input("Enter prime number q: "))
```

```python
print(p)
print(q)
```

Calculate n and phi(n)

```python
n=p*q
phi=(p-1)*(q-1)

print("n =",n)
print("phi(n) =",phi)
```

Find GCD Function

```python
def gcd(a,b):

    while b!=0:
        temp=b
        b=a%b
        a=temp

    return a
```

Choose Public Key e

```python
e=2

while e<phi:

    if gcd(e,phi)==1:
        break

    e=e+1

print("Public Key e =",e)
```

Calculate Private Key d

```python
d=1

while True:

    if (d*e)%phi==1:
        break

    d=d+1

print("Private Key d =",d)
```

Input Plaintext Message

```python
M=int(input("Enter plaintext message: "))
```

Encrypt Plaintext

```python
C=1

for i in range(e):
    C=(C*M)%n

print("Ciphertext =",C)
```

Decrypt Ciphertext

```python
P=1

for i in range(d):
    P=(P*C)%n

print("Decrypted Plaintext =",P)
```
