Keyword Input

```python
keyword=input("Enter keyword: ")
```

Convert Keyword to Uppercase and Remove Duplicates

```python
new_keyword=""

for char in keyword:
    if char>='a' and char<='z':
        char=chr(ord(char)-32)

    if char=='J':
        char='I'

    found=False

    for k in new_keyword:
        if k==char:
            found=True
            break

    if found==False and char>='A' and char<='Z':
        new_keyword=new_keyword+char

print("Keyword:",new_keyword)
```

Generate Playfair Key String

```python
alphabet="ABCDEFGHIKLMNOPQRSTUVWXYZ"

for char in alphabet:
    found=False

    for k in new_keyword:
        if k==char:
            found=True
            break

    if found==False:
        new_keyword=new_keyword+char
```

Create 5×5 Matrix

```python
matrix=[]
index=0

for i in range(5):
    row=[]

    for j in range(5):
        row.append(new_keyword[index])
        index=index+1

    matrix.append(row)
```

Display Matrix

```python
print("Playfair Key Matrix")

for i in range(5):
    for j in range(5):
        print(matrix[i][j],end=" ")
    print()
```

Plaintext Input

```python
plaintext=input("Enter plaintext: ")
```

Convert Plaintext to Uppercase and Remove Spaces

```python
text=""

for char in plaintext:
    if char>='a' and char<='z':
        char=chr(ord(char)-32)

    if char=='J':
        char='I'

    if char!=' ':
        text=text+char

print("Processed Text:",text)
```

Prepare Plaintext

```python
prepared=""
i=0

while i<len(text):

    first=text[i]

    if i+1<len(text):
        second=text[i+1]

        if first==second:
            prepared=prepared+first+"X"
            i=i+1
        else:
            prepared=prepared+first+second
            i=i+2

    else:
        prepared=prepared+first+"X"
        i=i+1

if len(prepared)%2!=0:
    prepared=prepared+"X"

print("Prepared Text:",prepared)
```

Show Pairs

```python
print("Pairs:")

for i in range(0,len(prepared),2):
    print(prepared[i]+prepared[i+1])
```

Find Position Function

```python
def position(letter):

    for row in range(5):
        for col in range(5):

            if matrix[row][col]==letter:
                return row,col
```

Encryption

```python
ciphertext=""

for i in range(0,len(prepared),2):

    a=prepared[i]
    b=prepared[i+1]

    row1,col1=position(a)
    row2,col2=position(b)

    if row1==row2:

        ciphertext=ciphertext+matrix[row1][(col1+1)%5]
        ciphertext=ciphertext+matrix[row2][(col2+1)%5]

    elif col1==col2:

        ciphertext=ciphertext+matrix[(row1+1)%5][col1]
        ciphertext=ciphertext+matrix[(row2+1)%5][col2]

    else:

        ciphertext=ciphertext+matrix[row1][col2]
        ciphertext=ciphertext+matrix[row2][col1]

print("Encrypted Message:",ciphertext)
```

Decryption

```python
decrypted=""

for i in range(0,len(ciphertext),2):

    a=ciphertext[i]
    b=ciphertext[i+1]

    row1,col1=position(a)
    row2,col2=position(b)

    if row1==row2:

        decrypted=decrypted+matrix[row1][(col1-1)%5]
        decrypted=decrypted+matrix[row2][(col2-1)%5]

    elif col1==col2:

        decrypted=decrypted+matrix[(row1-1)%5][col1]
        decrypted=decrypted+matrix[(row2-1)%5][col2]

    else:

        decrypted=decrypted+matrix[row1][col2]
        decrypted=decrypted+matrix[row2][col1]

print("Decrypted Message:",decrypted)
```
