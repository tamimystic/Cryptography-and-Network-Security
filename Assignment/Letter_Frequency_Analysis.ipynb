Write a program to perform letter frequency analysis on a ciphertext. <br>

Your program should: <br>

Count the frequency of each letter A–Z. <br>
Display the frequency count. <br>
Display the percentage frequency of each letter. <br>
Identify the top 5 most frequent letters. <br>

# ciphertext input

```python
ciphertext=input("Enter ciphertext: ").upper()
```

# create alphabet and count list

```python
alphabet="ABCDEFGHIJKLMNOPQRSTUVWXYZ"
frequency=[0]*26
```

# count letter frequency

```python
for char in ciphertext:
    if char.isalpha():
        index=alphabet.find(char)
        frequency[index]=frequency[index]+1
```

# frequency count

```python
print("Letter Frequency Count")

for i in range(26):
    print(alphabet[i],":",frequency[i])
```

# percentage frequency

```python
total_letters=sum(frequency)

print("Percentage Frequency")

for i in range(26):
    percentage=(frequency[i]/total_letters)*100
    print(alphabet[i],":",round(percentage,2),"%")
```

# find top 5

```python
letters=[]

for i in range(26):
    letters.append([frequency[i],alphabet[i]])

letters.sort(reverse=True)

print("Top 5 Most Frequent Letters")

for i in range(5):
    print(letters[i][1],":",letters[i][0])
```
