# AES S-Box Generation for One Byte

## Tasks

1. Take one byte as input in hexadecimal format
2. Find the multiplicative inverse of the byte in GF(2⁸)
3. Use the AES irreducible polynomial
4. Generate the AES S-Box value

Input Byte

```python
hex_byte=input("Enter a byte in hexadecimal: ")
hex_byte
```

Convert Hexadecimal to Integer

```python
byte=int(hex_byte,16)

print("Input Byte:",hex(byte))
```

AES Irreducible Polynomial

```python
irreducible=0x11B

print("AES Irreducible Polynomial:",hex(irreducible))
```

Galois Field Multiplication Function

```python
def gf_multiply(a,b):

    result=0

    while b>0:

        if b%2==1:
            result=result^a

        a=a<<1

        if a&0x100:
            a=a^0x11B

        b=b>>1

    return result&0xFF
```

Find Multiplicative Inverse

```python
inverse=0

if byte!=0:

    for i in range(1,256):

        value=gf_multiply(byte,i)

        if value==1:
            inverse=i
            break

print("Multiplicative Inverse:",hex(inverse))
```

Affine Transformation

```python
result=0

for i in range(8):

    bit=((inverse>>i)&1)^\
        ((inverse>>((i+4)%8))&1)^\
        ((inverse>>((i+5)%8))&1)^\
        ((inverse>>((i+6)%8))&1)^\
        ((inverse>>((i+7)%8))&1)^\
        ((0x63>>i)&1)

    result=result|(bit<<i)
```

Display AES S-Box Value

```python
print("AES S-Box Value:",hex(result))
```
