```python
def mod_inverse(a,m=26):
    for i in range(1,26):
        if (a*i)%m==1:
            return i
    return None
```

```python
def encrypt(text,a,b):
    result =""
    for ch in text:
        if ch.isalpha():
            start = ord('a') if ch.islower() else ord('A')
            result+=chr((((ord(ch)-start)*a) + b) % 26 +start)
        else:
            result+=ch
    return result
```

```python
a_inverse = mod_inverse(5)
a_inverse
```

```python
def decrypt(text,a,b):
    result =""
    for ch in text:
        if ch.isalpha():
            start = ord('a') if ch.islower() else ord('A')
            result+= chr(a_inverse*((ord(ch)-start)-b)%26 + start)
        else:
            result+=ch
    return result
```

```python
message = "Hello World"
a, b = 5, 8   # valid key

encrypted = encrypt(message, a, b)
decrypted = decrypt(encrypted, a, b)

print("Original :", message)
print("Encrypted:", encrypted)
print("Decrypted:", decrypted)
```
