# Mono Alphabetic

## Encryption

```python
encrypt_map ={
    'a':'m',
    'b':'n',
    'c':'b',
    'd':'v',
    'e':'c',
    'f':'x',
    'g':'z',
    'h':'l',
    'i':'k',
    'j':'j',
    'k':'h',
    'l':'g',
    'm':'f',
    'n':'d',
    'o':'s',
    'p':'a',
    'q':'p',
    'r':'o',
    's':'i',
    't':'u',
    'u':'y',
    'v':'t',
    'w':'r',
    'x':'e',
    'y':'w',
    'z':'q'
}
```

```python
text="Tamim Hossain"
```

```python
string = ""
```

```python
for ch in text.lower():
    if ch in encrypt_map:
        string = string + encrypt_map[ch]
    else:
        string = string+ch
```

```python
string
```

## Decryption

```python
encrypt_map
```

```python
encrypt_map.items()
```

```python
decrypted_map = {}

for k,v in encrypt_map.items():
    decrypted_map[v]=k
```

```python
decrypted_map
```

```python
decrypt_string =""
```

```python
for i in string.lower():
    if i in decrypted_map:
        decrypt_string +=decrypted_map[i]
    else:
        decrypt_string+=i
```
