```python
def encrypt(text,key):
    result = ""
    for ch in text:
        if ch.isalpha():
            start = ord('a') if ch.islower() else ord('A')
            result += chr((ord(ch) - start + key)%26+start)
        else:
            result+=ch
    return result
```

```python
def decrypt(text, key):
    return encrypt(text, -key)
```

```python
msg = "Hello World"
```

```python
enc_msg = encrypt(msg,3)
dec_msg = decrypt(enc_msg,3)
```
